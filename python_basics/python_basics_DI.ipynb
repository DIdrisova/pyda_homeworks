## Задание 1

Даны 2 переменных, в которых хранятся строки произвольной длины: *phrase_1* и *phrase_2*.  
Напишите код, который проверяет какая из этих строк длиннее.

Примеры работы программы:  
1.  

```
phrase_1 = 'Насколько проще было бы писать программы, если бы не заказчики'
phrase_2 = '640Кб должно хватить для любых задач. Билл Гейтс (по легенде)'

```

Результат:  
`Фраза 1 длиннее фразы 2`
2.     

```
phrase_1 = '640Кб должно хватить для любых задач. Билл Гейтс (по легенде)'
phrase_2 = 'Насколько проще было бы писать программы, если бы не заказчики'
```

Результат:  
`Фраза 2 длиннее фразы 1`
3.     

```
phrase_1 = 'Насколько проще было бы писать программы, если бы не заказчики'
phrase_2 = 'Насколько проще было бы писать программы, если бы не заказчики'
```

Результат:  
`Фразы равной длины`

````python
#Задание 1. Решение: 
phrase_1 = 'Насколько проще было бы писать программы, если бы не заказчики'
phrase_2 = '640Кб должно хватить для любых задач. Билл Гейтс (по легенде)'

if len(phrase_1)>len(phrase_2):
    print('Фраза 1 длиннее фразы 2')
elif len(phrase_1)<len(phrase_2):
    print('Фраза 2 длиннее фразы 1')
else: print('Фразы равной длины')
````

## Задание 2

Дана переменная, в которой хранится четырехзначное число (год). Необходимо написать программу, которая выведет, является ли данный год високосным или обычным.

Пример работы программы:
1.  

```
year = 2020
```
Результат:  
`Високосный год` 

2.  

```
year = 2019
```
Результат:  
`Обычный год` 

````python
#Задание 2. Решение: 
import datetime as dt
year=int(input('Введите год '))


if not(1<=year<=9998): print('Введите год от 1 до 9998')
else:   
    a=dt.date(year,1,1)
    b=dt.date(year+1,1,1)
    d=(b-a).days
    if d>365:
        print('Високосный год')
    else: print('Обычный год')
````

## Задание 5 (необязательное)

Дана переменная, в которой хранится шестизначное число (номер проездного билета). Напишите программу, которая будет определять, является ли данный билет "счастливым". Билет считается счастливым, если сумма первых трех цифр совпадает с суммой последних трех цифр номера.   

Примеры работы программы:  
1. ```number = 123456```

Результат:  
`Неасчастливый билет`
2. ```number = 123321```

Результат:  
`Счастливый билет`


````python
#Задание 5. Решение
n=input('Номер билета: ')
if len(n) != 6: 
    print('Номер должен содержать ровно 6 цифр')
elif int(n[0])+int(n[1])+int(n[2])==int(n[3])+int(n[4])+int(n[5]):
    print('Счастливый билет')
else: print('Несчастливый билет')
````

## Задание 6 (необязательное)

Напишите программу, которая сможет вычислять площади трех фигур (круг, треугольник и прямоугольник).
Тип фигуры запрашиваем через пользовательский ввод, после чего делаем запрос характеристик фигуры:

- если пользователь выбрал круг, запрашиваем его радиус,
- если треугольник – длины трех его сторон;
- если прямоугольник – длины двух его сторон.

Пример работы программы:
1.  

```
Введите тип фигуры:
Круг

Введите радиус круга:
10
```
Результат:  
`Площадь круга: 314.16`

2.  

```
Введите тип фигуры:
Треугольник

Введите длину стороны A:
2

Введите длину стороны B:
2

Введите длину стороны C:
3
```
Результат:  
`Площадь треугольника: 1.98`

````python
#Задание 6: Решение
import math 
opt=int(input('Выберите вариант 1,2 или 3: 1 - круг, 2 - треугольник, 3 - прямоугольник: '))
if (opt<1 or opt>3):
    print('Вариант недоступен')
elif opt==1: 
        rad=int(input('Введите радиус круга: '))
        print('Площадь круга: ',math.pi*(rad**2))
elif opt==2: 
        a=int(input('Введите длину стороны A: '))
        b=int(input('Введите длину стороны B: '))
        c=int(input('Введите длину стороны C: '))
        # для вычисления по формуле Герона вычислим полупериметр p:
        p=(a+b+c)/2
        print('Площадь треугольника: ',math.sqrt(p*(p-a)*(p-b)*(p-c)))
elif opt==3: 
        l=int(input('Введите длину стороны A: '))
        h=int(input('Введите длину стороны B: '))
        print('Площадь прямоугольника: ',l*h)
````
