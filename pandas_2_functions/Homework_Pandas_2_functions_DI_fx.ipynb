# Задание 1**
Напишите функцию, которая классифицирует фильмы из материалов занятия по следующим правилам:
- оценка 2 и меньше - низкий рейтинг
- оценка 4 и меньше - средний рейтинг
- оценка 4.5 и 5 - высокий рейтинг

Результат классификации запишите в столбец class

```python
import pandas as pd
```

```python
# открываю файл с рейтингами - смотрю
df = pd.read_csv('ml-latest-small/ratings.csv')
df.head()
```

```python
def grouping(rating):
    """
    функция определяет группу по рейтингу
    """
    if rating<=2:
        group = 'низкий рейтинг'
    elif rating<=4:
        group = 'средний рейтинг'
    else:
        group = 'средний рейтинг'
    return group
```

```python
#ghbvtyztv функцию к столбцу рейтинг - смортим результат
df['class']=df['rating'].apply(grouping)
df.head()
```

# Задание 2
Используем файл keywords.csv.

Необходимо написать гео-классификатор, который каждой строке сможет выставить географическую принадлежность определенному региону. Т. е. если поисковый запрос содержит название города региона, то в столбце ‘region’ пишется название этого региона. Если поисковый запрос не содержит названия города, то ставим ‘undefined’.

Правила распределения по регионам Центр, Северо-Запад и Дальний Восток:

geo_data = {
'Центр': ['москва', 'тула', 'ярославль'],

'Северо-Запад': ['петербург', 'псков', 'мурманск'],

'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
}

Результат классификации запишите в отдельный столбец region.

```python
# открываю файл с рейтингами - смотрю
df = pd.read_csv('ml-latest-small/keywords.csv')
```

```python
# смотрю, какие года есть в файле - необязательный шаг
df['keyword'].unique()
```

```python
# определяю словарь geo_data
geo_data = { 'Центр': ['москва', 'тула', 'ярославль'],
'Северо-Запад': ['петербург', 'псков', 'мурманск'],
'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск'] }
geo_data
```

```python
def identify_region(keyword):
    """
    функция определяет регион по ключевым словам
    """
    region = 0
    for geo_region, cities in geo_data.items():
        for city in cities:
            if city.lower() in keyword.lower():
                region = geo_region
                pass
            
    if region==0:
        region = 'undefined'
    return region
```

```python
# применяю функцию к столбцу рейтинг - смортю результат
df['region']=df['keyword'].apply(identify_region)
df.head()
```

```python
# проверка для ключевых слов, которые совпадают с названием города
import numpy as np
all_kw = list(np.concatenate(list(geo_data.values())))
filtered_df = df[df['keyword'].isin(all_kw)]

filtered_df.head()
```

# Задание 3 (бонусное)

Есть мнение, что “раньше снимали настоящее кино, не то что сейчас”. Ваша задача проверить это утверждение, используя файлы с рейтингами фильмов из прошлого домашнего занятия (файл movies.csv и ratings.csv из базы https://grouplens.org/datasets/movielens). Т. е. проверить верно ли, что с ростом года выпуска фильма его средний рейтинг становится ниже.

При этом мы не будем затрагивать субьективные факторы выставления этих рейтингов, а пройдемся по следующему алгоритму:

В переменную years запишите список из всех годов с 1950 по 2010.

Напишите функцию production_year, которая каждой строке из названия фильма выставляет год выпуска. Не все названия фильмов содержат год выпуска в одинаковом формате, поэтому используйте следующий алгоритм:

для каждой строки пройдите по всем годам списка years
если номер года присутствует в названии фильма, то функция возвращает этот год как год выпуска
если ни один из номеров года списка years не встретился в названии фильма, то возвращается 1900 год
Запишите год выпуска фильма по алгоритму пункта 2 в новый столбец ‘year’

Посчитайте средний рейтинг всех фильмов для каждого значения столбца ‘year’ и отсортируйте результат по убыванию рейтинга

```python
years = range(1950,2010,1)
```

```python
# открываю файл с рейтингами - смотрю
df = pd.read_csv('ml-latest-small/movies.csv')
df.head()
```

```python
def production_year(title):
    """
    функция проходит по всем годам списка year:
    - если номер года присутствует в названии фильма,
    то функция возвращает этот год как год выпуска
    - если ни один из номеров года списка years не встретился в названии фильма, 
    то возвращается 1900 год
    """
    result = 0
    for year in years:
        if str(year) in title:
            result = year
    if result == 0:
        result = 1900
    return result
```

```python
# применяю фунцкцию к колонке title - смотрю начало
df['year']=df['title'].apply(production_year)
df.head()
```

```python
# на всякий случай смотрю конец датафрейма. Есть фильмы, вышедшие после 2010 и до 1950 
# т.к. они не входят в переменную years, возвращается 1900, что удовлетворяет условиям задачи
df.tail(10)
```

```python
# оставляю только необходимую информацию в новом датафрейме
movie_year = df.filter(items = ['movieId','year'])
movie_year.head(10)
```

```python
# открываю файл с рейтингами, подтягиваю год создания по movieId
ratings = pd.read_csv('ml-latest-small/ratings.csv')
ratings_year = ratings.merge(movie_year, how='left', on='movieId')
ratings_year.head()
```

```python
# группирую по годам, считаю среднее, соритриую по убыванию рейтинга, оставляю только нужные колонки - год и рейтинг
ratings_year.groupby('year').mean().sort_values('rating', ascending=False).filter(['year','rating'])
```
